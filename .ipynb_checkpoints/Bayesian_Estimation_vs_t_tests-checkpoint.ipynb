```python
import arviz as az
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pymc as pm

from scipy import stats 

print(f"Running on PyMC3 v{pm.__version__}")
```

```python
%config InlineBackend.figure_format = 'retina'
az.style.use("arviz-darkgrid")
rng = np.random.default_rng(seed=42)
```

# Problem
We want to evaluate the clinical trial for a drug. The drug claim that it is suppose to increase intelligence by comparing IQ scores of individuals. Control recieves a placebo and Treatment recieves the drug

```python
# data
drug = (101,100,102,104,102,97,105,105,98,101,100,123,105,103,100,95,102,106,
        109,102,82,102,100,102,102,101,102,102,103,103,97,97,103,101,97,104,
        96,103,124,101,101,100,101,101,104,100,101)
placebo = (99,101,100,101,102,100,97,101,104,101,102,102,100,105,88,101,100,
           104,100,100,100,101,102,103,97,101,101,100,101,99,101,100,100,
           101,100,99,101,100,102,99,100,99)


y1 = np.array(drug)
y2 = np.array(placebo)
y = pd.DataFrame(
    dict(value=np.r_[y1, y2], group=np.r_[["drug"] * len(drug), ["placebo"] * len(placebo)])
)
y
```

```python
# visualization
y.hist("value", by="group", figsize=(12, 4));
```

```python
# Priors
Î¼_m = y.value.mean()
Î¼_s = y.value.std() * 2
Ïƒ_low = 1
Ïƒ_high = 10

with pm.Model() as model:
    group1_mean = pm.Normal("group1_mean", mu=Î¼_m, sigma=Î¼_s)
    group2_mean = pm.Normal("group2_mean", mu=Î¼_m, sigma=Î¼_s)


with model:
    group1_std = pm.Uniform("group1_std", lower=Ïƒ_low, upper=Ïƒ_high)
    group2_std = pm.Uniform("group2_std", lower=Ïƒ_low, upper=Ïƒ_high)

with model:
    Î½ = pm.Exponential("Î½_minus_one", 1 / 29.0)
```

```python
pm.model_to_graphviz(model)
```

```python
# prior for  Î½
az.plot_kde(rng.exponential(scale=30, size=10000), fill_kwargs={"alpha": 0.5});
```

```python
# Likelihood
with model:
    Î»1 = group1_std ** -2
    Î»2 = group2_std ** -2

    # Liklihood
    group1 = pm.StudentT("drug", nu=Î½, mu=group1_mean, lam=Î»1, observed=y1)
    group2 = pm.StudentT("placebo", nu=Î½, mu=group2_mean, lam=Î»2, observed=y2)

    
```

```python
pm.model_to_graphviz(model)
```

```python
with model:
    diff_of_means = pm.Deterministic("difference of means", group1_mean - group2_mean)
    diff_of_stds = pm.Deterministic("difference of stds", group1_std - group2_std)
    effect_size = pm.Deterministic(
        "effect size", diff_of_means / np.sqrt((group1_std ** 2 + group2_std ** 2) / 2)
    )
```

```python
pm.model_to_graphviz(model)
```

```python
with model:
    trace = pm.sample(2000, return_inferencedata=True)
```

```python
az.plot_posterior(
    trace,
    var_names=["group1_mean", "group2_mean", "group1_std", "group2_std", "Î½_minus_one"],
    color="#87ceeb",
);
```

```python
az.plot_posterior(
    trace,
    var_names=["difference of means", "difference of stds", "effect size"],
    ref_val=0,
    color="#87ceeb",
);
```

```python
az.plot_forest(trace, var_names=["group1_mean", "group2_mean"]);
```

```python
az.plot_forest(trace, var_names=["group1_std", "group2_std", "Î½_minus_one"]);
```

```python
az.summary(trace, var_names=["difference of means", "difference of stds", "effect size"])
```

# T-Test

```python
# When compared to the t-test, what observation can it give
# Running the t-test
t, p = stats.ttest_ind(drug, placebo)
print('Drug vs. Placebo t = ', str(t), ' p = ', str(p))
```

# Conclusion
As we can see here, if we run a t-test on the same data, it gives a p-value of 0.12. According to industry standards, people are following 0.10 or 0.05 as a threshold. A statistican would have made a decision about the result being inconclusive based on the t-test.

While doing Bayesian Estimation, we observed that the difference in means is above 0 (97% of the time). This suggest that some significant difference exist between two groups. One can estimate that the drug have a significant effect on their IQ after consuming it.
